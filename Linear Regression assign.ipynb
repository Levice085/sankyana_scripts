## **Predicting Housing Prices**
### Problem Statement:
The goal of this project is to build a linear regression model to predict house prices based 
on various features such as house age, distance to the nearest MRT station, number of 
convenience stores, and other relevant attributes. The model will help in understanding 
the impact of different factors on house prices of unit area and assist potential buyers and 
real estate agents in making informed decisions.

### Data Exploration and cleaning

```python
import pandas as pd
import numpy as np
```

```python
#Importing dataset
house_data = pd.read_csv(r'C:\Users\hp\Desktop\5.1\sankyana\Assignment\ML project 1\Real estate.csv')
```

```python
house_data.head()
```

```python
#Checking for missing values
house_data.isnull().sum()
```

```python
#Summary of the data
house_data.describe().T
```

```python
house_data.columns
house_data.info()
```

```python
#Defining the dependent and independent variables
Y = house_data['Y house price of unit area']
X = house_data.drop(['Y house price of unit area'], axis = 1)
```

```python
#Checking the shape of the dependent and independent variables
print(X.shape)
print(Y.shape)
```

### Data spliting

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

### Building and training the model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
#Training the model
model.fit(x_train, y_train)
```

```python
#Predicting the model
y_pred = model.predict(x_test)
y_pred
```

### Evaluating the model

```python
from sklearn.metrics import mean_squared_error as mse, r2_score
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
#Printing the mean squared error
print("The mean squared is %.3f" % mse(y_test, y_pred))
```

```python
# Printing the R2 score
print("The R2 score is %.3f" % r2_score(y_test,y_pred))
```
