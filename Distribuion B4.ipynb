```python
mu = 42
std = 10
snd = stats.norm(mu, std)
```

```python
## Plot a normal distribution graph using the generated random values with the mean 42, std 10

plt.figure(figsize=(7.5, 7.5))
plt.plot (x_random, snd.pdf(x_random)) # pdf - notifyies my code that I am trying to plot a graph using the randomlly generated values

# defining the x-axis limit
plt.xlim(-10,100)
# title
plt.title("Normala distribution Graph, mean=42, std=10", fontsize= 15)
plt.xlabel("Values for the Random Variable",  fontsize= 15)
plt.ylabel("Probability", fontsize= 15)
```

### Business Scenario - Bernoullis distribution 


Create a Bar plot of Bernoullis distribution where;

It should contain the Instance of Bernoulli distribution with parameter p = 0.7
Outcome of experiment can take value as 0, 1
Use pmf function to determine the probability of different values of random variable

This bar plot must contain the following :

It should have the x-axis limit of (-1, 2)

The plot color should be lime and green.

The figure size should be (7, 7).
The plot should contain title and labels on x and y axis with font size=15.

Note: Use the Python Scipy Bernoulli class, to calculate the probability mass function values(pmf).


```python
# importing scipy Bernoulli class
from scipy.stats import bernoulli
X = [0,1]# random variable

p=0.7 # probability of success
plt.figure(figsize = (7,7))
plt.bar(X, bernoulli.pmf(X, p), width = 0.3, color = ['lime' ,'green'] )
plt.title("Bernouli Distribution Graph for the probability of success as p=0.7", fontsize= 15)
plt.xlabel("Values random variable X= [0 to 1]",  fontsize= 15)
plt.ylabel("Probability of Success", fontsize= 15)
```

### Business Scenario - Poisson Distribution 


To find probability equal to some value -

Create a new python code to generate Poisson distribution by importing poisson from scipy.stats library and calculate the pmf, when a store sells 3 apples per day on average. What is the probability that they will sell 5 apples on a given day?


```python
# importing poison
from scipy.stats import poisson

# calculating the probability of selling 5 apples in a day using PMF

PEqual= poisson.pmf(mu = 3, k=5)
print("Propability of seling 5 apples in a day is:",PEqual)
```

 Create a new python code to generate Poisson distribution and Calculate the cumulative distribution function (cdf), when a certain player kicks seven goals per match on average. What is the probability that this player will kick four or less goals in a given match?


```python
# calculating the probability of selling 5 apples in a day using cdf

PEqual1= poisson.cdf(mu =7 , k=4)
print("Propability of kicking 4 or less goal is a match:",PEqual1)
```
